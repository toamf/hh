```python
import random
from scapy.all import Ether, Dot1Q, IP, TCP, wrpcap
from scapy.contrib.mpls import MPLS

def random_ip():
    return ".".join(map(str, (random.randint(0, 255) for _ in range(4))))

def random_port():
    return random.randint(1024, 65535)

def random_mac():
    return ":".join(["{:02x}".format(random.randint(0, 255)) for _ in range(6)])

def create_packet():
    packet = (
        Ether(dst=random_mac(), src=random_mac(), type=0x8100) / 
        Dot1Q(vlan=random.randint(1, 4095), prio=random.randint(0, 7)) / 
        MPLS(label=random.randint(0, 1048575), s=1, ttl=random.randint(1, 255), cos=random.randint(0, 7)) / 
        IP(src=random_ip(), dst=random_ip(), ttl=random.randint(1, 255), tos=random.randint(0, 255)) / 
        TCP(sport=random_port(), dport=random_port(), seq=random.randint(0, 4294967295), ack=random.randint(0, 4294967295), flags=random.randint(0, 31))
    )
    return packet

packets = [create_packet() for _ in range(100)]

wrpcap("output.pcap", packets)

print("OK!")
```

```python
from scapy.all import rdpcap

packets = rdpcap("output.pcap")

for packet in packets:
    print(packet.summary())
```
